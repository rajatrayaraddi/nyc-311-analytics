## Data Cleaning/Preprocessing

```python
#Standardizing column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
print(df.columns[:10])
```

```python
# Converting date columns to datetime format
date_cols = ["created_date", "closed_date", "due_date", "resolution_action_updated_date"]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(
            df[col],
            format="%m/%d/%Y %I:%M:%S %p",
            errors="coerce"
        )
df.head()
```

```python
# Extracting date features from 'created_date'
df["year"] = df["created_date"].dt.year
df["month"] = df["created_date"].dt.month
df["dayofweek"] = df["created_date"].dt.dayofweek
df["hour"] = df["created_date"].dt.hour
df.head()
```

```python
# Converting coordinate columns to numeric
coord_cols = [
    "x_coordinate_(state_plane)",
    "y_coordinate_(state_plane)",
    "latitude",
    "longitude"
]

for col in coord_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")
df.head()
```

```python
# Calculating 'days_to_close' if 'closed_date' exists
if "closed_date" in df.columns:
    df["days_to_close"] = (df["closed_date"] - df["created_date"]).dt.days
df.head()
```

```python
# Selecting relevant columns for analysis
keep_cols = [
    "unique_key", "created_date", "closed_date", "year", "month", "dayofweek", "hour",
    "agency", "complaint_type", "descriptor", "borough", "location_type", "city", "incident_address", "street_name",
    "incident_zip", "latitude", "longitude", "days_to_close", "status", "location"
]

df_clean = df[keep_cols].copy()
df_clean.head()
```

```python
# Handling missing values
cat_unknown_cols = [
    "descriptor",
    "location_type",
    "city",
    "incident_address",
    "street_name"
]

for col in cat_unknown_cols:
    df_clean[col] = df_clean[col].fillna("Unknown")

df_clean.isna().sum()
```

```python
# Dropping rows with missing critical values
df_clean = df_clean.dropna(subset=["latitude", "longitude", "incident_zip"])
df_clean.isna().sum()
```

```python
# Standardizing 'incident_zip' to first 5 digits
if "incident_zip" in df_clean.columns:
    df_clean["incident_zip"] = (
        df_clean["incident_zip"]
        .astype(str)
        .str.extract(r"(\d+)", expand=False)
        .str[:5]
    )
df_clean.head()
```

```python
df_clean.to_csv("nyc311_1m_clean.csv", index=False)
print("Saved cleaned dataset: nyc311_1m_clean.csv")
```
